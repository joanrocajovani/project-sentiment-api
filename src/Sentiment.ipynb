```python
import pandas as pd
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("../src/tableclean.csv")
```

```python
df.head()
```

```python
nltk.download('stopwords')
stop = stopwords.words('english')
```

```python
def words_filter(df_,new_column, old_column, stopwords):
    df_[new_column] = df_[old_column].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
    return
```

```python
words_filter(df, "Overview_2", "Overview", stop)
```

```python
df.head()
```

```python
nltk.downloader.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
df[['neg', 'neu', 'pos', 'compound']] = df['Overview_2'].apply(sid.polarity_scores).apply(pd.Series)
```

```python
df
```

```python
df.to_csv('tableclean2.csv', encoding='utf-8', index = False)
```
